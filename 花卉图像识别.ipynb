### 制作好数据源：
- data_transforms中指定了所有图像预处理操作
- ImageFolder假设所有的文件按文件夹保存好，每个文件夹下面存贮同一类别的图片，文件夹的名字为分类的名字

````python
data_transforms = {
    # 按顺序去做接下来的每一步操作
    'train': transforms.Compose([
        transforms.Resize([96, 96]),# 将图片更改大小，过小会丢失态度信息，过大会运行过慢
        # 数据增强
        transforms.RandomRotation(45),#随机旋转，-45到45度之间随机选
        transforms.CenterCrop(224),#从中心开始裁剪
        transforms.RandomHorizontalFlip(p=0.5),#随机水平翻转 选择一个概率概率
        transforms.RandomVerticalFlip(p=0.5),#随机垂直翻转
        # 下面这两行用的少
        transforms.ColorJitter(brightness=0.2, contrast=0.1, saturation=0.1, hue=0.1),#参数1为亮度，参数2为对比度，参数3为饱和度，参数4为色相
        transforms.RandomGrayscale(p=0.025),#概率转换成灰度率，3通道就是R=G=B
        transforms.ToTensor(), # 把数据转换为tensor，也就是torch中支持的一种解构
        # 标准化操作，从已有的模型训练数据中得到相应的数据
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])#均值，标准差
    ]),
    'valid': transforms.Compose([transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        # 测试集中也进行相同的标准化操作
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
}
````

````python
batch_size = 128

image_datasets = {x: datasets.ImageFolder(os.path.join(data_dir, x), data_transforms[x]) for x in ['train', 'valid']}
dataloaders = {x: torch.utils.data.DataLoader(image_datasets[x], batch_size=batch_size, shuffle=True) for x in ['train', 'valid']}
dataset_sizes = {x: len(image_datasets[x]) for x in ['train', 'valid']}
class_names = image_datasets['train'].classes
````

### 加载models中提供的模型，并且直接用训练的好权重当做初始化参数
- 第一次执行需要下载，可能会比较慢，我会提供给大家一份下载好的，可以直接放到相应路径

````python
# 直接利用resnet网络模型。
# 方法一：直接从torch中拉下来；（torch已经收录了很多经典的网络模型）
# 方法二：直接将配置文件复制粘贴过来
model_name = 'resnet'  #可选的比较多 ['resnet', 'alexnet', 'vgg', 'squeezenet', 'densenet', 'inception']
#是否用人家训练好的特征来做，true的时候，所有的参数都不进行更新，都“冻住”
feature_extract = True 
````

通用代码（不必修改），选择是在CPU上还是GPU上训练

````python
# 这段是通用代码，不必修改
# 是否用GPU训练
train_on_gpu = torch.cuda.is_available()

if not train_on_gpu:
    print('CUDA is not available.  Training on CPU ...')
else:
    print('CUDA is available!  Training on GPU ...')
    
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
````

定义模型，选用xx层的resnet，这里可以直接**通过.拿到**，之后的模型需要自己去定义

````python
# model是torchvision中导入的一个包
# 选用18层的resnet较快，条件好的可以选用152层的，还有101、50层等
model_ft = models.resnet18()
# 会打印出来所有的网络结构
model_ft

# 参数解析（这里是别人的模型）：
# 第一层Conv2d：2d的卷积，输入通道为3，输出通道为64，stride(height, width),第一层的kernel会比较大，能够将图片在第一层就压缩的比较小
# 每一层之后以batchNorm进行归一化
# 最后做了一个全局的平均池化，无论输入多大，都可以得到一个固定的输出
````

````python
# model要传resnet
def set_parameter_requires_grad(model, feature_extracting):
    if feature_extracting:
        for param in model.parameters():
            # 表明反向传播时是否计算梯度，没有梯度那么参数就会保持不变
            param.requires_grad = False
````

把模型输出层改成自己的

````python
def initialize_model(model_name, num_classes, feature_extract, use_pretrained=True):
    # 选择合适的模型，不同模型的初始化方法稍微有点区别
    model_ft = models.resnet18(pretrained=use_pretrained)
    # 设定所有参数都不更新了
    set_parameter_requires_grad(model_ft, feature_extract)

    # 修改全连接层为自己的
    num_ftrs = model_ft.fc.in_features
    model_ft.fc = nn.Linear(num_ftrs, num_classes)

    input_size = 64

    return model_ft, input_size
````

设置哪些层需要训练

````python
# 初始化，执行上面定义的初始化模型函数
model_ft, input_size = initialize_model(model_name, 102, feature_extract, use_pretrained=True)

#GPU计算
model_ft = model_ft.to(device)

# 保存模型训练结果：网络结构图、所有的权重参数等
filename='checkpoint.pth'

# 是否训练所有层，存了权重参数和偏置参数
params_to_update = model_ft.parameters()

print("Params to learn:")
if feature_extract:
    # 默认置空，需要的时候再传进来
    params_to_update = []
    for name,param in model_ft.named_parameters():
        # 如果需要梯度，也就是参数进行更新
        if param.requires_grad == True:
            # 将梯度保存进去
            params_to_update.append(param)
            print("\t",name)
else:
    for name,param in model_ft.named_parameters():
        if param.requires_grad == True:
            print("\t",name)
````

优化器设置

````python
# 优化器设置
# Adam是现在用的很多的也很合适的优化器，params_to_update中放的是需要更新的参数，学习率指定为0.01
optimizer_ft = optim.Adam(params_to_update, lr=1e-2)
# 定义“学习率的衰减策略”，参数：优化器是什么、执行几次优化器衰减一次学习率、学习率变为原来的多少
scheduler = optim.lr_scheduler.StepLR(optimizer_ft, step_size=7, gamma=0.1)#学习率每7个epoch衰减成原来的1/10
#最后一层已经LogSoftmax()了，所以不能nn.CrossEntropyLoss()来计算了，nn.CrossEntropyLoss()相当于logSoftmax()和nn.NLLLoss()整合
# 交叉熵，损失函数
criterion = nn.NLLLoss()
````

### 训练模块

````python
def train_model(model, dataloaders, criterion, optimizer, num_epochs=25,filename=filename):
    # 得到当前的时间
    since = time.time()
    # 记录模型最好的准确率
    best_acc = 0
    # 进行判断，决定把模型放到gpu/cpu中
    model.to(device)

    # 训练过程中打印一堆损失和指标
    val_acc_history = []
    train_acc_history = []
    train_losses = []
    valid_losses = []

    # 学习率
    LRs = [optimizer.param_groups[0]['lr']]

    # 记录效果最好的模型，先对其进行初始化
    best_model_wts = copy.deepcopy(model.state_dict())

    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)

        # 训练和验证
        for phase in ['train', 'valid']:
            if phase == 'train':
                model.train()  # 训练
            else:
                model.eval()   # 验证，验证集中不涉及到参数更新

            # 初始化每次的损失和准确率
            running_loss = 0.0
            running_corrects = 0

            # 把数据都取个遍
            for inputs, labels in dataloaders[phase]:
                # 将数据标签放到cpu/gpu中
                inputs = inputs.to(device)
                labels = labels.to(device)

                # 梯度清零（每次执行的时候要先进行梯度清零
                optimizer.zero_grad()
                # 只有训练的时候计算和更新梯度
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                # 占位符  preds预测值
                _, preds = torch.max(outputs, 1)

                # 训练阶段更新权重（只有训练阶段才进行更新）
                if phase == 'train':
                    # 损失的反向传播
                    loss.backward()
                    # 参数的更新
                    optimizer.step()

                # 计算损失
                running_loss += loss.item() * inputs.size(0)  # inputs.size()相当于是batch，running_loss是当前这一次迭代总的损失
                running_corrects += torch.sum(preds == labels.data)   # 保存正确结果的个数

            # 迭代完一个epoch算平均值，损失、准确率
            epoch_loss = running_loss / len(dataloaders[phase].dataset)
            epoch_acc = running_corrects.double() / len(dataloaders[phase].dataset)
            
            # 打印一个epoch浪费了多少时间
            time_elapsed = time.time() - since
            print('Time elapsed {:.0f}m {:.0f}s'.format(time_elapsed // 60, time_elapsed % 60))
            # 打印损失和准确率
            print('{} Loss: {:.4f} Acc: {:.4f}'.format(phase, epoch_loss, epoch_acc))
            

            # 得到最好那次的模型
            if phase == 'valid' and epoch_acc > best_acc:
                best_acc = epoch_acc
                # 更新记录的当前最优权重值
                best_model_wts = copy.deepcopy(model.state_dict())
                state = {
                  'state_dict': model.state_dict(),
                  'best_acc': best_acc,
                  'optimizer' : optimizer.state_dict(),
                }
                torch.save(state, filename)  # 保存到本地

            if phase == 'valid':
                val_acc_history.append(epoch_acc)
                valid_losses.append(epoch_loss)
            if phase == 'train':
                train_acc_history.append(epoch_acc)
                train_losses.append(epoch_loss)
        
        # 打印学习率
        print('Optimizer learning rate : {:.7f}'.format(optimizer.param_groups[0]['lr']))
        LRs.append(optimizer.param_groups[0]['lr'])  # 保存当前学习率
        print()
        scheduler.step()  # 学习率衰减

    time_elapsed = time.time() - since  # 总体花费的时间
    print('Training complete in {:.0f}m {:.0f}s'.format(time_elapsed // 60, time_elapsed % 60))
    print('Best val Acc: {:4f}'.format(best_acc))  # 打印出验证集中最好的效果

    # 训练完后用最好的一次当做模型最终的结果
    model.load_state_dict(best_model_wts)
    return model, val_acc_history, train_acc_history, valid_losses, train_losses, LRs 
````

### 测试数据预处理

- 测试数据处理方法需要跟训练时一直才可以 
- crop操作的目的是保证输入的大小是一致的
- 标准化操作也是必须的，用跟训练数据相同的mean和std,但是需要注意一点训练数据是在0-1上进行标准化，所以测试数据也需要先归一化
- 最后一点，PyTorch中颜色通道是第一个维度，跟很多工具包都不一样，需要转换


````python
def process_image(image_path):
    # 读取测试数据
    img = Image.open(image_path)
    # Resize,thumbnail方法只能进行缩小，所以进行了判断
    if img.size[0] > img.size[1]:
        img.thumbnail((10000, 256))
    else:
        img.thumbnail((256, 10000))
    # Crop操作
    left_margin = (img.width-224)/2
    bottom_margin = (img.height-224)/2
    right_margin = left_margin + 224
    top_margin = bottom_margin + 224
    img = img.crop((left_margin, bottom_margin, right_margin,   
                      top_margin))
    # 相同的预处理方法
    img = np.array(img)/255
    mean = np.array([0.485, 0.456, 0.406]) #provided mean
    std = np.array([0.229, 0.224, 0.225]) #provided std
    img = (img - mean)/std
    
    # 注意颜色通道应该放在第一个位置
    img = img.transpose((2, 0, 1))
    
    return img
````

````python
def imshow(image, ax=None, title=None):
    """展示数据"""
    if ax is None:
        fig, ax = plt.subplots()
    
    # 颜色通道还原
    image = np.array(image).transpose((1, 2, 0))
    
    # 预处理还原
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    image = std * image + mean
    image = np.clip(image, 0, 1)
    
    ax.imshow(image)
    ax.set_title(title)
    
    return ax
````
